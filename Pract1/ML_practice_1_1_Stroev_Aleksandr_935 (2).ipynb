````python
import pandas as pd
import numpy as np

# 1. –û—Ç–∫—Ä—ã—Ç–∏–µ —Ñ–∞–π–ª–∞ –∏ –≤—ã–≤–æ–¥ 3 –ø–æ—Å–ª–µ–¥–Ω–∏—Ö —Å—Ç—Ä–æ–∫
df = pd.read_excel('_end_seminar.xlsx', engine='openpyxl')
print("3 –ø–æ—Å–ª–µ–¥–Ω–∏—Ö —Å—Ç—Ä–æ–∫–∏:")
print(df.tail(3))

# –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö (–¥–ª—è —Ä–∞–∑–º—ã—à–ª–µ–Ω–∏—è):
# - 347 —Å—Ç—Ä–æ–∫ (–ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –¥—É–±–ª–∏–∫–∞—Ç–æ–≤)
# - –°—Ç–æ–ª–±—Ü—ã: timestamp, id, rating, 18_group, is_mi, fall_1-fall_3, spring_1-spring_3,
#   is_first_time, percentile, 17_group, blended, is_ml_student
# - –î–∞–Ω–Ω—ã–µ: —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ–ø—Ä–æ—Å–∞ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –æ –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏—è—Ö –ø–æ –≤—ã–±–æ—Ä—É –∫—É—Ä—Å–æ–≤

# 2. –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –∏–∑ –≥—Ä—É–ø–ø 18 –∏ 17 –≥–æ–¥–∞
# –ì—Ä—É–ø–ø—ã 18 –≥–æ–¥–∞: –∫–æ–ª–æ–Ω–∫–∞ '18_group' –Ω–µ –ø—É—Å—Ç–∞—è
# –ì—Ä—É–ø–ø—ã 17 –≥–æ–¥–∞: –∫–æ–ª–æ–Ω–∫–∞ '17_group' –Ω–µ –ø—É—Å—Ç–∞—è
users_18 = df['id'][df['18_group'].notna()].nunique()
users_17 = df['id'][df['17_group'].notna()].nunique()
print(f"\n–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑ –≥—Ä—É–ø–ø 18 –≥–æ–¥–∞: {users_18}")
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑ –≥—Ä—É–ø–ø 17 –≥–æ–¥–∞: {users_17}")

# 3. –°—Ç—É–¥–µ–Ω—Ç—ã —Å —Ä–∞–≤–Ω—ã–º–∏ –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è–º–∏
duplicates_percentile = df.duplicated(subset=['percentile'], keep=False).sum()
print(f"\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å –Ω–µ—É–Ω–∏–∫–∞–ª—å–Ω—ã–º–∏ –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è–º–∏: {duplicates_percentile}")
print(f"–ï—Å—Ç—å –ª–∏ —Å—Ç—É–¥–µ–Ω—Ç—ã —Å —Ä–∞–≤–Ω—ã–º–∏ –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—è–º–∏? {'–î–∞' if duplicates_percentile > 0 else '–ù–µ—Ç'}")

# 4. –ü–æ–∏—Å–∫ –ø—Ä–æ–ø—É—Å–∫–æ–≤
missing = df.isna().sum()
missing_cols = missing[missing > 0]
print("\n–ü—Ä–æ–ø—É—Å–∫–∏ –≤ –∫–æ–ª–æ–Ω–∫–∞—Ö:")
print(missing_cols)

# 5. –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –±–µ–∑ –ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–∏—è –≤—Å–µ—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
# –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ç–∏–ø—ã –∫–æ–ª–æ–Ω–æ–∫
string_cols = df.select_dtypes(include=['object']).columns
numeric_cols = df.select_dtypes(include=['number']).columns
bool_cols = df.select_dtypes(include=['bool']).columns

# –°–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é –¥–ª—è –±–µ–∑–æ–ø–∞—Å–Ω–æ–π —Ä–∞–±–æ—Ç—ã
df_clean = df.copy()

# –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ç–∏–ø–∞
for col in string_cols:
    df_clean[col] = df_clean[col].fillna('')

for col in numeric_cols:
    df_clean[col] = df_clean[col].fillna(0)

for col in bool_cols:
    df_clean[col] = df_clean[col].fillna(False)

# –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π –≤–µ–∫—Ç–æ—Ä–∏–∑–æ–≤–∞–Ω–Ω—ã–π –ø–æ–¥—Ö–æ–¥ –±–µ–∑ —è–≤–Ω—ã—Ö —Ü–∏–∫–ª–æ–≤ –ø–æ –∫–æ–ª–æ–Ω–∫–∞–º:
# df_clean = df.fillna({
#     col: '' if df[col].dtype == 'object'
#     else (False if df[col].dtype == 'bool' else 0)
#     for col in df.columns[df.isna().any()]
# })

# 6. –ê–Ω–∞–ª–∏–∑ –∫–æ–ª–æ–Ω–∫–∏ 'is_first_time'
print("\n–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ 'is_first_time':")
print(df_clean['is_first_time'].value_counts(dropna=False))

# 7. –£–¥–∞–ª–µ–Ω–∏–µ –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –æ–±—Ä–∞—â–µ–Ω–∏–π (–æ—Å—Ç–∞–≤–ª—è–µ–º —Å–∞–º—É—é –ø–æ–∑–¥–Ω—é—é –≤–µ—Ä—Å–∏—é)
# –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –∏ —É–¥–∞–ª—è–µ–º –¥—É–±–ª–∏–∫–∞—Ç—ã –ø–æ id, –æ—Å—Ç–∞–≤–ª—è—è –ø–æ—Å–ª–µ–¥–Ω–µ–µ
df_clean = df_clean.sort_values('timestamp').drop_duplicates(subset='id', keep='last')
print(f"\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –¥—É–±–ª–∏–∫–∞—Ç–æ–≤: {len(df_clean)}")

# 8. –ü—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –æ—Ç–≤–µ—Ç–æ–≤ –≤ 'is_first_time'
total = len(df_clean)
no_count = (df_clean['is_first_time'] == '–ù–µ—Ç').sum()
yes_count = (df_clean['is_first_time'] == '–î–∞').sum()
no_percent = no_count / total * 100
print(f"\n–ü—Ä–æ—Ü–µ–Ω—Ç –æ—Ç–≤–µ—Ç–æ–≤ '–ù–µ—Ç': {no_percent:.2f}%")
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–≤–µ—Ç–æ–≤ '–ù–µ—Ç': {no_count}")

# 9. –ê–Ω–∞–ª–∏–∑ blended-–∫—É—Ä—Å–æ–≤ –¥–ª—è —á–µ—Ç–≤–µ—Ä–æ–∫—É—Ä—Å–Ω–∏–∫–æ–≤ (17-–π –≥–æ–¥)
blended_17 = df_clean[df_clean['17_group'].notna() & (df_clean['blended'] != '')]
blended_courses = blended_17['blended'].value_counts()
top_course = blended_courses.idxmax()
top_count = blended_courses.max()

# –°—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –ø–æ –∫—É—Ä—Å–∞–º
avg_rating = blended_17.groupby('blended')['rating'].mean()
best_course = avg_rating.idxmax()
best_course_students = blended_17[blended_17['blended'] == best_course].shape[0]

print(f"\nBlended-–∫—É—Ä—Å—ã –¥–ª—è —á–µ—Ç–≤–µ—Ä–æ–∫—É—Ä—Å–Ω–∏–∫–æ–≤: {blended_courses.index.tolist()}")
print(f"–ö—É—Ä—Å —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Å—Ç—É–¥–µ–Ω—Ç–æ–≤: {top_course} ({top_count} —Å—Ç—É–¥–µ–Ω—Ç–æ–≤)")
print(f"–ö—É—Ä—Å —Å —Å–∞–º—ã–º –≤—ã—Å–æ–∫–∏–º —Å—Ä–µ–¥–Ω–∏–º —Ä–µ–π—Ç–∏–Ω–≥–æ–º: {best_course} ({best_course_students} —Å—Ç—É–¥–µ–Ω—Ç–æ–≤)")

# 10. –°—Ç—É–¥–µ–Ω—Ç—ã —Å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º–∏ –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏—è–º–∏ –ø–æ –≤—Å–µ–º –∫—É—Ä—Å–∞–º
# –§–æ—Ä–º–∏—Ä—É–µ–º –∫–ª—é—á –∏–∑ –≤—Å–µ—Ö –∫—É—Ä—Å–æ–≤—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
course_cols = ['fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3', 'blended']
# –ó–∞–º–µ–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ blended –Ω–∞ –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É –¥–ª—è 3-–∫—É—Ä—Å–Ω–∏–∫–æ–≤
df_clean['blended_filled'] = df_clean['blended'].fillna('')

# –°–æ–∑–¥–∞–µ–º —Å–æ—Å—Ç–∞–≤–Ω–æ–π –∫–ª—é—á
df_clean['course_set'] = df_clean[course_cols].astype(str).agg('|'.join, axis=1)

# –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –Ω–∞ –∫–∞–∂–¥—ã–π –Ω–∞–±–æ—Ä
course_set_counts = df_clean['course_set'].value_counts()
duplicates = course_set_counts[course_set_counts > 1]

print(f"\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –∫—É—Ä—Å–æ–≤ —Å –¥—É–±–ª–∏–∫–∞—Ç–∞–º–∏: {len(duplicates)}")
print("–ü—Ä–∏–º–µ—Ä—ã –Ω–∞–±–æ—Ä–æ–≤ —Å –¥—É–±–ª–∏–∫–∞—Ç–∞–º–∏ (–ø–µ—Ä–≤—ã–µ 5):")
print(duplicates.head())

# 11. –û–±—â–∏–µ –∫—É—Ä—Å—ã –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ 17 –∏ 18 –≥–æ–¥–∞
# –°–æ–±–∏—Ä–∞–µ–º –≤—Å–µ –∫—É—Ä—Å—ã –∏–∑ –≤—Å–µ—Ö –∫–æ–ª–æ–Ω–æ–∫ –∫—É—Ä—Å–æ–≤
all_course_cols = ['fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3', 'blended']

# –ö—É—Ä—Å—ã 18 –≥–æ–¥–∞
courses_18 = set()
for col in all_course_cols:
    courses_18.update(df_clean[df_clean['18_group'].notna()][col].dropna().unique())

# –ö—É—Ä—Å—ã 17 –≥–æ–¥–∞
courses_17 = set()
for col in all_course_cols:
    courses_17.update(df_clean[df_clean['17_group'].notna()][col].dropna().unique())

common_courses = courses_18.intersection(courses_17) - {''}
print(f"\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—â–∏—Ö –∫—É—Ä—Å–æ–≤: {len(common_courses)}")
print(f"–û–±—â–∏–µ –∫—É—Ä—Å—ã: {sorted(common_courses)}")

# 12. –ö—É—Ä—Å—ã —Ç–æ–ª—å–∫–æ –¥–ª—è 18 –∏ —Ç–æ–ª—å–∫–æ –¥–ª—è 17 –≥–æ–¥–∞
only_18 = courses_18 - courses_17 - {''}
only_17 = courses_17 - courses_18 - {''}

print(f"\n–ö—É—Ä—Å—ã —Ç–æ–ª—å–∫–æ –¥–ª—è 18 –≥–æ–¥–∞ ({len(only_18)}): {sorted(only_18)}")
print(f"–ö—É—Ä—Å—ã —Ç–æ–ª—å–∫–æ –¥–ª—è 17 –≥–æ–¥–∞ ({len(only_17)}): {sorted(only_17)}")
````

````python
import pandas as pd
import numpy as np

# –ó–∞–≥—Ä—É–∑–∫–∞ –∏ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
df = pd.read_excel('_end_seminar.xlsx', engine='openpyxl')

# –£–¥–∞–ª—è–µ–º –¥—É–±–ª–∏–∫–∞—Ç—ã, –æ—Å—Ç–∞–≤–ª—è—è —Å–∞–º—É—é –ø–æ–∑–¥–Ω—é—é –∑–∞–ø–∏—Å—å
df = df.sort_values('timestamp').drop_duplicates(subset='id', keep='last').reset_index(drop=True)

# –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏
df = df.fillna({
    col: '' if df[col].dtype == 'object'
    else (False if df[col].dtype == 'bool' else 0)
    for col in df.columns
})

# === –®–ê–ì 1: –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Å–µ–Ω–Ω–∏—Ö –∫—É—Ä—Å–æ–≤ (–í–ï–ö–¢–û–†–ò–ó–û–í–ê–ù–ù–û) ===
# –°–æ–∑–¥–∞–µ–º –∫–æ–ª–æ–Ω–∫—É —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫—É—Ä—Å–æ–≤, –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ 1 –¥–ª—è –≤—Å–µ—Ö
df['num_fall_courses'] = 1

# –°—Ç—É–¥–µ–Ω—Ç—ã 17 –≥–æ–¥–∞: –ú–û–ü (171, 172) –∏ –¢–ò (173) ‚Üí 2 –∫—É—Ä—Å–∞
# –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º 17_group –∫–∞–∫ —Å—Ç—Ä–æ–∫—É –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–≥–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
mask_17_mop_ti = (
    (df['17_group'].notna()) &
    (df['17_group'] != '') &
    (df['17_group'].astype(str).str[:3].isin(['171', '172', '173']))
)
df.loc[mask_17_mop_ti, 'num_fall_courses'] = 2

# –°—Ç—É–¥–µ–Ω—Ç—ã 18 –≥–æ–¥–∞: –ú–ò (is_mi == 1) ‚Üí 2 –∫—É—Ä—Å–∞
mask_18_mi = (df['18_group'].notna()) & (df['18_group'] != '') & (df['is_mi'] == 1)
df.loc[mask_18_mi, 'num_fall_courses'] = 2

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –∏–∑ –∑–∞–¥–∞–Ω–∏—è
assert (df[df['id'] == '2662600c2c37e11e62f6ee0b88452f22']['num_fall_courses'] == 2).all(), \
    "–û—à–∏–±–∫–∞ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ 2662600c2c37e11e62f6ee0b88452f22"
assert (df[df['id'] == 'd555d2805e1d93d4f023e57dc4c8f403']['num_fall_courses'] == 2).all(), \
    "–û—à–∏–±–∫–∞ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ d555d2805e1d93d4f023e57dc4c8f403"
assert (df[df['id'] == '8fe79f84f36e3a5d2d6745621321302c']['num_fall_courses'] == 1).all(), \
    "–û—à–∏–±–∫–∞ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ 8fe79f84f36e3a5d2d6745621321302c"
assert (df[df['id'] == 'e4caca755ee0bdd711e18fb8084958b5']['num_fall_courses'] == 1).all(), \
    "–û—à–∏–±–∫–∞ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–∞ e4caca755ee0bdd711e18fb8084958b5"

print("–ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–æ–π–¥–µ–Ω–∞! –í—Å–µ —Å—Ç—É–¥–µ–Ω—Ç—ã –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω—ã.")
print(f"–°—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å 1 –∫—É—Ä—Å–æ–º: {(df['num_fall_courses'] == 1).sum()}")
print(f"–°—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å 2 –∫—É—Ä—Å–∞–º–∏: {(df['num_fall_courses'] == 2).sum()}")

# === –®–ê–ì 2: –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è ===
# –°–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
dist_df = df[['id', 'num_fall_courses', 'percentile', 'fall_1', 'fall_2', 'fall_3']].copy()

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–æ–ª–æ–Ω–∫–∏ –¥–ª—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
dist_df['course1'] = '???'
dist_df['course2'] = '-'

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –ª–∏–º–∏—Ç—ã –º–µ—Å—Ç –Ω–∞ –∫—É—Ä—Å–∞—Ö
course_limits = {
    'Statistical Learning Theory': 60,
    '–í—ã—Å–æ–∫–æ–ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω—ã–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è': 60,
    '–ê–Ω–∞–ª–∏–∑ –Ω–µ—Å—Ç—Ä—É–∫—Ç—É—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö': 60,  # –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º 60 –∫–∞–∫ —É –¥—Ä—É–≥–∏—Ö –∏—Å–∫–ª—é—á–µ–Ω–∏–π
    # –í—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ –∫—É—Ä—Å—ã ‚Äî 30 –º–µ—Å—Ç –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é
}

# –°–æ–±–∏—Ä–∞–µ–º –≤—Å–µ —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∫—É—Ä—Å—ã –∏–∑ –¥–∞–Ω–Ω—ã—Ö
all_courses = set()
for col in ['fall_1', 'fall_2', 'fall_3']:
    all_courses.update(dist_df[col].unique())
all_courses = {c for c in all_courses if c != '' and pd.notna(c)}

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫–∏ –∑–∞–Ω—è—Ç—ã—Ö –º–µ—Å—Ç
occupied = {course: 0 for course in all_courses}
limit = {course: course_limits.get(course, 30) for course in all_courses}

# === –®–ê–ì 3: –ê–ª–≥–æ—Ä–∏—Ç–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ –≤–æ–ª–Ω–∞–º ===

# –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–π –≤–æ–ª–Ω–µ
def distribute_wave(priority_cols, occupied, limit):
    """
    –†–∞—Å–ø—Ä–µ–¥–µ–ª—è–µ—Ç —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –Ω–∞ –∫—É—Ä—Å–∞—Ö —Å–æ–≥–ª–∞—Å–Ω–æ —É–∫–∞–∑–∞–Ω–Ω—ã–º –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω—ã–º –∫–æ–ª–æ–Ω–∫–∞–º.

    priority_cols: —Å–ø–∏—Å–æ–∫ –∫–æ–ª–æ–Ω–æ–∫ —Å –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞–º–∏ –¥–ª—è —ç—Ç–æ–π –≤–æ–ª–Ω—ã (–Ω–∞–ø—Ä–∏–º–µ—Ä, ['fall_1', 'fall_2'])
    occupied: —Å–ª–æ–≤–∞—Ä—å —Å —Ç–µ–∫—É—â–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–Ω—è—Ç—ã—Ö –º–µ—Å—Ç –Ω–∞ –∫–∞–∂–¥–æ–º –∫—É—Ä—Å–µ
    limit: —Å–ª–æ–≤–∞—Ä—å —Å –ª–∏–º–∏—Ç–∞–º–∏ –º–µ—Å—Ç –Ω–∞ –∫–∞–∂–¥–æ–º –∫—É—Ä—Å–µ

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ–±–Ω–æ–≤–ª–µ–Ω–Ω—ã–π —Å–ª–æ–≤–∞—Ä—å occupied
    """
    # –°–æ–∑–¥–∞–µ–º –≤—Ä–µ–º–µ–Ω–Ω—É—é —Ç–∞–±–ª–∏—Ü—É —Å –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞–º–∏ –¥–ª—è —ç—Ç–æ–π –≤–æ–ª–Ω—ã
    wave_data = []
    for idx, row in dist_df.iterrows():
        # –ü—Ä–æ–ø—É—Å–∫–∞–µ–º —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —É–∂–µ –ø–æ–ª—É—á–∏–ª–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫—É—Ä—Å–æ–≤
        courses_assigned = (row['course1'] != '???') + (row['course2'] != '-' and row['course2'] != '???')
        if courses_assigned >= row['num_fall_courses']:
            continue

        # –î–æ–±–∞–≤–ª—è–µ–º –≤—Å–µ –∫—É—Ä—Å—ã –∏–∑ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫ –¥–ª—è —ç—Ç–æ–≥–æ —Å—Ç—É–¥–µ–Ω—Ç–∞
        for col in priority_cols:
            course = row[col]
            if course != '' and pd.notna(course) and occupied.get(course, 0) < limit.get(course, 30):
                wave_data.append({
                    'idx': idx,
                    'id': row['id'],
                    'course': course,
                    'percentile': row['percentile']
                })

    # –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—é (–ø–æ —É–±—ã–≤–∞–Ω–∏—é ‚Äî –ª—É—á—à–∏–µ —Å—Ç—É–¥–µ–Ω—Ç—ã –ø–µ—Ä–≤—ã–º–∏)
    wave_df = pd.DataFrame(wave_data).sort_values('percentile', ascending=False)

    # –†–∞—Å–ø—Ä–µ–¥–µ–ª—è–µ–º —Å—Ç—É–¥–µ–Ω—Ç–æ–≤
    for _, record in wave_df.iterrows():
        course = record['course']
        idx = record['idx']

        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –µ—â–µ —Å–≤–æ–±–æ–¥–Ω—ã–µ –º–µ—Å—Ç–∞
        if occupied.get(course, 0) >= limit.get(course, 30):
            continue

        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —Å–∫–æ–ª—å–∫–æ –∫—É—Ä—Å–æ–≤ —É–∂–µ –ø–æ–ª—É—á–∏–ª —Å—Ç—É–¥–µ–Ω—Ç
        courses_assigned = (dist_df.loc[idx, 'course1'] != '???') + \
                          (dist_df.loc[idx, 'course2'] != '-' and dist_df.loc[idx, 'course2'] != '???')

        # –ï—Å–ª–∏ —Å—Ç—É–¥–µ–Ω—Ç—É –Ω—É–∂–Ω–æ –±–æ–ª—å—à–µ –∫—É—Ä—Å–æ–≤ ‚Äî –∑–∞—á–∏—Å–ª—è–µ–º
        if courses_assigned < dist_df.loc[idx, 'num_fall_courses']:
            if dist_df.loc[idx, 'course1'] == '???':
                dist_df.loc[idx, 'course1'] = course
            elif dist_df.loc[idx, 'course2'] == '-':
                dist_df.loc[idx, 'course2'] = course
            else:
                dist_df.loc[idx, 'course2'] = course

            occupied[course] = occupied.get(course, 0) + 1

    return occupied

# –ü–µ—Ä–≤–∞—è –≤–æ–ª–Ω–∞:
# - –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å 1 –∫—É—Ä—Å–æ–º: —Ç–æ–ª—å–∫–æ fall_1
# - –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å 2 –∫—É—Ä—Å–∞–º–∏: fall_1 –∏ fall_2 (–æ–±–∞ —Å—á–∏—Ç–∞—é—Ç—Å—è –ø–µ—Ä–≤—ã–º –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–æ–º)
occupied = distribute_wave(['fall_1', 'fall_2'], occupied, limit)

# –í—Ç–æ—Ä–∞—è –≤–æ–ª–Ω–∞:
# - –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å 1 –∫—É—Ä—Å–æ–º: fall_2
# - –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å 2 –∫—É—Ä—Å–∞–º–∏: fall_3
occupied = distribute_wave(['fall_2', 'fall_3'], occupied, limit)

# –¢—Ä–µ—Ç—å—è –≤–æ–ª–Ω–∞ (—Ç–æ–ª—å–∫–æ –¥–ª—è —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å 1 –∫—É—Ä—Å–æ–º, –∫–æ—Ç–æ—Ä—ã–º –Ω–µ —Ö–≤–∞—Ç–∏–ª–æ –º–µ—Å—Ç–∞ –Ω–∞ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –≤–æ–ª–Ω–∞—Ö):
# - –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å 1 –∫—É—Ä—Å–æ–º: fall_3
occupied = distribute_wave(['fall_3'], occupied, limit)

# === –®–ê–ì 4: –§–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ –∏—Ç–æ–≥–æ–≤–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ ===
result_df = dist_df[['id', 'course1', 'course2']].copy()

# –î–ª—è —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–º –Ω—É–∂–µ–Ω —Ç–æ–ª—å–∫–æ 1 –∫—É—Ä—Å, —É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º course2 = '-'
mask_single = df.set_index('id').loc[result_df['id'], 'num_fall_courses'].values == 1
result_df.loc[mask_single, 'course2'] = '-'

# –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
result_df.to_csv('res_fall.csv', index=False)

print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–∞–≤–µ—Ä—à–µ–Ω–æ!")
print(f"–í—Å–µ–≥–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤: {len(result_df)}")
print(f"–°—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å 1 –∫—É—Ä—Å–æ–º: {(result_df['course2'] == '-').sum()}")
print(f"–°—Ç—É–¥–µ–Ω—Ç–æ–≤ —Å 2 –∫—É—Ä—Å–∞–º–∏: {(result_df['course2'] != '-').sum()}")
print(f"–°—Ç—É–¥–µ–Ω—Ç–æ–≤, –Ω–µ –ø–æ–ª—É—á–∏–≤—à–∏—Ö –Ω–∏ –æ–¥–Ω–æ–≥–æ –∫—É—Ä—Å–∞: {(result_df['course1'] == '???').sum()}")
print(f"–°—Ç—É–¥–µ–Ω—Ç–æ–≤, –ø–æ–ª—É—á–∏–≤—à–∏—Ö 1 –∫—É—Ä—Å –∏–∑ 2 –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö: {((result_df['course1'] != '???') & (result_df['course2'] == '???')).sum()}")

# –ü—Ä–∏–º–µ—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ –∫—É—Ä—Å–∞–º
print("\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫—É—Ä—Å–∞–º (–æ—Å–µ–Ω–Ω–∏–µ):")
for course in sorted(occupied.keys()):
    print(f"  {course}: {occupied[course]}/{limit[course]} –º–µ—Å—Ç")
````

### üìå –¶–µ–ª—å –∑–∞–Ω—è—Ç–∏—è
–û—Å–≤–æ–∏—Ç—å –±–∞–∑–æ–≤—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–∞—Ä–∏–π –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω—ã—Ö –≤ Python, –≤—ã–ø–æ–ª–Ω–∏—Ç—å —Ä–∞–∑–≤–µ–¥–æ—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑ (Exploratory Data Analysis, EDA) –Ω–∞ –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ Iris –∏ —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å —à–∞–±–ª–æ–Ω –Ω–æ—É—Ç–±—É–∫–∞ –¥–ª—è –ø–æ—Å–ª–µ–¥—É—é—â–∏—Ö —Ä–∞–±–æ—Ç. –û—Å–æ–±–æ–µ –≤–Ω–∏–º–∞–Ω–∏–µ ‚Äî –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏–∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–π.

### üìö –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è —Å–ø—Ä–∞–≤–∫–∞

**–î–∞—Ç–∞—Å–µ—Ç Iris** ‚Äî –∏—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∞—è ¬´–≤–∏–∑–∏—Ç–Ω–∞—è –∫–∞—Ä—Ç–æ—á–∫–∞¬ª –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è. –í 1936 –≥–æ–¥—É –±—Ä–∏—Ç–∞–Ω—Å–∫–∏–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫ –†–æ–Ω–∞–ª—å–¥ –§–∏—à–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª –µ–≥–æ –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ –ª–∏–Ω–µ–π–Ω–æ–≥–æ –¥–∏—Å–∫—Ä–∏–º–∏–Ω–∞–Ω—Ç–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞. –î–∞—Ç–∞—Å–µ—Ç —Å–æ–¥–µ—Ä–∂–∏—Ç 150 –æ–±—Ä–∞–∑—Ü–æ–≤ —Ç—Ä—ë—Ö –≤–∏–¥–æ–≤ –∏—Ä–∏—Å–æ–≤ (*Iris setosa*, *Iris versicolor*, *Iris virginica*) —Å —á–µ—Ç—ã—Ä—å–º—è –º–æ—Ä—Ñ–æ–ª–æ–≥–∏—á–µ—Å–∫–∏–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏:
- `sepal length` ‚Äî –¥–ª–∏–Ω–∞ —á–∞—à–µ–ª–∏—Å—Ç–∏–∫–∞ (—Å–º)
- `sepal width` ‚Äî —à–∏—Ä–∏–Ω–∞ —á–∞—à–µ–ª–∏—Å—Ç–∏–∫–∞ (—Å–º)
- `petal length` ‚Äî –¥–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞ (—Å–º)
- `petal width` ‚Äî —à–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞ (—Å–º)

````python
# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–Ω–∞—á–µ–Ω–∏—è –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª –î–û –∏–º–ø–æ—Ä—Ç–∞ –±–∏–±–ª–∏–æ—Ç–µ–∫, –≤–ª–∏—è—é—â–∏—Ö –Ω–∞ —Å–ª—É—á–∞–π–Ω–æ—Å—Ç—å
import numpy as np
np.random.seed(42)  # –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø—Ä–∏ –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –∑–∞–ø—É—Å–∫–∞—Ö
````

````python
from sklearn.datasets import load_iris
import pandas as pd

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
iris = load_iris()
# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤ —É–¥–æ–±–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç —Å –∏–º–µ–Ω–∞–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target  # –¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è (—á–∏—Å–ª–æ–≤—ã–µ –º–µ—Ç–∫–∏)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)  # –ß–µ–ª–æ–≤–µ–∫–æ—á–∏—Ç–∞–µ–º—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è
````

#### –®–∞–≥ 3. –ü–µ—Ä–≤–∏—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã


````python
# –ë—ã—Å—Ç—Ä—ã–π –ø—Ä–æ—Å–º–æ—Ç—Ä –ø–µ—Ä–≤—ã—Ö 5 —Å—Ç—Ä–æ–∫
print(df.head())

# –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ç–∏–ø–∞—Ö –¥–∞–Ω–Ω—ã—Ö –∏ –ø—Ä–æ–ø—É—Å–∫–∞—Ö
print(df.info())

# –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: —Å—Ä–µ–¥–Ω–µ–µ, –°–ö–û, –∫–≤–∞—Ä—Ç–∏–ª–∏
print(df.describe())
````


#### –®–∞–≥ 4. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤–∑–∞–∏–º–æ—Å–≤—è–∑–µ–π




````python

import matplotlib.pyplot as plt
import seaborn as sns

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Å—Ç–∏–ª—è
sns.set(style="whitegrid", palette="Set2")

# –ü–∞—Ä–Ω—ã–µ –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ—è–Ω–∏—è —Å —Ü–≤–µ—Ç–æ–≤—ã–º —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ–º –ø–æ –≤–∏–¥–∞–º
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='species', diag_kind='hist', markers=['o', 's', 'D'])
plt.suptitle('–ü–∞—Ä–Ω—ã–µ –¥–∏–∞–≥—Ä–∞–º–º—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏—Ä–∏—Å–æ–≤', y=1.02, fontsize=14)
plt.show()

# –¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π
plt.figure(figsize=(8, 6))
sns.heatmap(df.iloc[:, :4].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤')
plt.show()
````

````python
# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –≤—Å–µ—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å –Ω–∞–ª–æ–∂–µ–Ω–∏–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø–æ –≤–∏–¥–∞–º
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
for idx, col in enumerate(iris.feature_names):
    ax = axes[idx // 2, idx % 2]
    sns.histplot(data=df, x=col, hue='species', kde=True, ax=ax, bins=15)
    ax.set_title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ: {col}')
plt.tight_layout()
plt.show()
````

#### –®–∞–≥ 6. –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é


````python
# X ‚Äî –º–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–≤—Å–µ —Å—Ç–æ–ª–±—Ü—ã –∫—Ä–æ–º–µ —Ü–µ–ª–µ–≤—ã—Ö)
X = df.drop(['target', 'species'], axis=1).values
# y ‚Äî –≤–µ–∫—Ç–æ—Ä –º–µ—Ç–æ–∫ (–º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∫–∞–∫ —á–∏—Å–ª–æ–≤—ã–µ, —Ç–∞–∫ –∏ —Å—Ç—Ä–æ–∫–æ–≤—ã–µ)
y = df['target'].values  # –∏–ª–∏ df['species'].values –¥–ª—è —Å—Ç—Ä–æ–∫–æ–≤—ã—Ö –º–µ—Ç–æ–∫

print(f"–†–∞–∑–º–µ—Ä –º–∞—Ç—Ä–∏—Ü—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ X: {X.shape}")
print(f"–†–∞–∑–º–µ—Ä –≤–µ–∫—Ç–æ—Ä–∞ –º–µ—Ç–æ–∫ y: {y.shape}")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º seed –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏
np.random.seed(42)

# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# ========== 1. –ü–µ—Ä–≤–∏—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö ==========
print("–ü–ï–†–í–´–ï 5 –°–¢–†–û–ö:")
print(df.head())
print("\n–ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –î–ê–ù–ù–´–•:")
df.info()
print("\n–û–ü–ò–°–ê–¢–ï–õ–¨–ù–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê:")
print(df.describe())

# ========== 2. –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ X –∏ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é y ==========
X = df[iris.feature_names]
y = df['species']
print("\n–†–ê–ó–î–ï–õ–ï–ù–ò–ï –ù–ê X –ò y:")
print(f"–†–∞–∑–º–µ—Ä X: {X.shape}, —Ä–∞–∑–º–µ—Ä y: {y.shape}")

# ========== 3. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è ‚Äì —è—â–∏–∫–∏ —Å —É—Å–∞–º–∏ ==========
sns.set_style("whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('–Ø—â–∏–∫–∏ —Å —É—Å–∞–º–∏ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ Iris –ø–æ –≤–∏–¥–∞–º', fontsize=16)

for i, ax in enumerate(axes.flat):
    feature = df.columns[i]
    sns.boxplot(x='species', y=feature, data=df, ax=ax)
    ax.set_title(feature)
    ax.set_xlabel('')
    ax.set_ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ (—Å–º)')
plt.tight_layout()
plt.show()

# –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
print("\n–ò–ù–¢–ï–†–ü–†–ï–¢–ê–¶–ò–Ø –ì–†–ê–§–ò–ö–û–í:")
print("- –î–ª–∏–Ω–∞ –∏ —à–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞ (petal length/width) —Ö–æ—Ä–æ—à–æ —Ä–∞–∑–¥–µ–ª—è—é—Ç –≤–∏–¥—ã: setosa –∏–º–µ–µ—Ç –º–∞–ª–µ–Ω—å–∫–∏–µ –ª–µ–ø–µ—Å—Ç–∫–∏, virginica ‚Äì –±–æ–ª—å—à–∏–µ.")
print("- –î–ª–∏–Ω–∞ —á–∞—à–µ–ª–∏—Å—Ç–∏–∫–∞ (sepal length) —Ç–∞–∫–∂–µ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è, –Ω–æ –µ—Å—Ç—å –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏–µ.")
print("- –®–∏—Ä–∏–Ω–∞ —á–∞—à–µ–ª–∏—Å—Ç–∏–∫–∞ (sepal width) —É –≤–∏–¥–æ–≤ —á–∞—Å—Ç–∏—á–Ω–æ –ø–µ—Ä–µ–∫—Ä—ã–≤–∞–µ—Ç—Å—è, —É virginica –∑–∞–º–µ—Ç–Ω—ã –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–µ –≤—ã–±—Ä–æ—Å—ã.")

# ========== 4. –ê–Ω–∞–ª–∏–∑ –≤—ã–±—Ä–æ—Å–æ–≤ –¥–ª—è –≤–∏–¥–∞ virginica –ø–æ —à–∏—Ä–∏–Ω–µ —á–∞—à–µ–ª–∏—Å—Ç–∏–∫–∞ ==========
target_species = 'virginica'
target_feature = 'sepal width (cm)'
virginica_data = df[df['species'] == target_species][target_feature]

Q1 = virginica_data.quantile(0.25)
Q3 = virginica_data.quantile(0.75)
IQR = Q3 - Q1
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR

outliers = virginica_data[(virginica_data < lower_whisker) | (virginica_data > upper_whisker)]
outlier_count = len(outliers)
outlier_percent = (outlier_count / len(virginica_data)) * 100

print("\n–†–ï–ó–£–õ–¨–¢–ê–¢–´ –ê–ù–ê–õ–ò–ó–ê –í–´–ë–†–û–°–û–í")
print(f"–ü—Ä–∏–∑–Ω–∞–∫: {target_feature}")
print(f"–í–∏–¥: {target_species}")
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—Ä–∞–∑—Ü–æ–≤: {len(virginica_data)}")
print(f"Q1: {Q1:.3f}, Q3: {Q3:.3f}, IQR: {IQR:.3f}")
print(f"–ì—Ä–∞–Ω–∏—Ü—ã: [{lower_whisker:.3f}, {upper_whisker:.3f}]")
print(f"–í—ã–±—Ä–æ—Å–æ–≤: {outlier_count} ({outlier_percent:.1f}%)")
if outlier_count > 0:
    print("–ó–Ω–∞—á–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤:", sorted(outliers.values))
else:
    print("–í—ã–±—Ä–æ—Å—ã –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç.")

print("\n–ò–ù–¢–ï–†–ü–†–ï–¢–ê–¶–ò–Ø: –£ –≤–∏–¥–∞ virginica –ø–æ —à–∏—Ä–∏–Ω–µ —á–∞—à–µ–ª–∏—Å—Ç–∏–∫–∞ –æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã –≤—ã–±—Ä–æ—Å—ã ‚Äì –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ü–≤–µ—Ç–∫–æ–≤ —Å –∞–Ω–æ–º–∞–ª—å–Ω–æ –Ω–∏–∑–∫–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏. –≠—Ç–æ –º–æ–∂–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ –≤–Ω—É—Ç—Ä–∏–≤–∏–¥–æ–≤—É—é –≤–∞—Ä–∏–∞—Ç–∏–≤–Ω–æ—Å—Ç—å –∏–ª–∏ –æ—à–∏–±–∫–∏ –∏–∑–º–µ—Ä–µ–Ω–∏–π.")
````

–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤–∞—Ä–∏–∞—Ü–∏–∏ (—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ / —Å—Ä–µ–¥–Ω–µ–µ √ó 100%) –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞. –ö–∞–∫–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –Ω–∞–∏–±–æ–ª–µ–µ –∏–∑–º–µ–Ω—á–∏–≤ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ —Å–≤–æ–µ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è? –ü–æ—á–µ–º—É —ç—Ç–æ –≤–∞–∂–Ω–æ –¥–ª—è –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

np.random.seed(42)

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# ==== 1. –ü–µ—Ä–≤–∏—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑ ====
print("–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫:")
print(df.head())
print("\n–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö:")
print(df.info())
print("\n–û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
print(df.describe())

# ==== 2. –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ X –∏ y ====
X = df[iris.feature_names]
y = iris.target  # –∏–ª–∏ iris.target_names[iris.target] –µ—Å–ª–∏ –Ω—É–∂–Ω–æ —Å—Ç—Ä–æ–∫–æ–≤—ã–µ –º–µ—Ç–∫–∏
print("\n–†–∞–∑–º–µ—Ä X:", X.shape)
print("–†–∞–∑–º–µ—Ä y:", y.shape)

# ==== 3. –†–∞—Å—á—ë—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –≤–∞—Ä–∏–∞—Ü–∏–∏ ====
print("\n–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤–∞—Ä–∏–∞—Ü–∏–∏ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ Iris:")
cv_dict = {}
for col in df.columns:
    mean = df[col].mean()
    std = df[col].std()
    cv = (std / mean) * 100
    cv_dict[col] = cv
    print(f"{col}: {cv:.2f}%")

# ==== 4. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –≤–∞—Ä–∏–∞—Ü–∏–∏ ====
plt.figure(figsize=(8, 5))
plt.bar(cv_dict.keys(), cv_dict.values(), color='skyblue', edgecolor='black')
plt.ylabel('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤–∞—Ä–∏–∞—Ü–∏–∏ (%)')
plt.title('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤–∞—Ä–∏–∞—Ü–∏–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ Iris')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# ==== 5. –ê–Ω–∞–ª–∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ ====
max_cv_feature = max(cv_dict, key=cv_dict.get)
max_cv_value = cv_dict[max_cv_feature]
print(f"\n–ù–∞–∏–±–æ–ª–µ–µ –∏–∑–º–µ–Ω—á–∏–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫: '{max_cv_feature}' ({max_cv_value:.2f}%)")
print("–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è: –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å –≤—ã—Å–æ–∫–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–º –≤–∞—Ä–∏–∞—Ü–∏–∏ –∏–º–µ—é—Ç –±–æ–ª—å—à–æ–π —Ä–∞–∑–±—Ä–æ—Å")
print("–æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ —Å—Ä–µ–¥–Ω–µ–≥–æ, —á—Ç–æ –≤–∞–∂–Ω–æ —É—á–∏—Ç—ã–≤–∞—Ç—å –ø—Ä–∏ –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö –ø–µ—Ä–µ–¥")
print("–ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤, —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã—Ö –∫ –º–∞—Å—à—Ç–∞–±—É (–Ω–∞–ø—Ä–∏–º–µ—Ä, SVM, kNN).")
````

–°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫ ¬´—Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –∏ —à–∏—Ä–∏–Ω—ã –ª–µ–ø–µ—Å—Ç–∫–∞¬ª (petal_length / petal_width). –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —ç—Ç–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –ø–æ –≤–∏–¥–∞–º. –£–ª—É—á—à–∞–µ—Ç –ª–∏ –æ–Ω —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

np.random.seed(42)

# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# ==== 1. –ü–µ—Ä–≤–∏—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö ====
print("–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫:")
print(df.head())
print("\n–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö:")
print(df.info())
print("\n–û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
print(df.describe())

# ==== 2. –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ X –∏ y ====
X = df[iris.feature_names]
y = df['species']
print("\n–†–∞–∑–º–µ—Ä X:", X.shape)
print("–†–∞–∑–º–µ—Ä y:", y.shape)

# ==== 3. –°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è ====
df['petal_ratio'] = df['petal length (cm)'] / df['petal width (cm)']

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –Ω–æ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
plt.figure(figsize=(10, 6))
for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.hist(subset['petal_ratio'], bins=10, alpha=0.5, label=species, edgecolor='black')
plt.xlabel('–û—Ç–Ω–æ—à–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –∫ —à–∏—Ä–∏–Ω–µ –ª–µ–ø–µ—Å—Ç–∫–∞')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—Ç–Ω–æ—à–µ–Ω–∏—è –¥–ª–∏–Ω–∞/—à–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞ –ø–æ –≤–∏–¥–∞–º –∏—Ä–∏—Å–æ–≤')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for species in df['species'].unique():
    subset = df[df['species'] == species]
    axes[0].hist(subset['petal length (cm)'], bins=10, alpha=0.5, label=species, edgecolor='black')
    axes[1].hist(subset['petal width (cm)'], bins=10, alpha=0.5, label=species, edgecolor='black')
axes[0].set_xlabel('–î–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞ (—Å–º)')
axes[0].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
axes[0].set_title('–î–ª–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞')
axes[0].legend()
axes[0].grid(True, linestyle='--', alpha=0.6)
axes[1].set_xlabel('–®–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞ (—Å–º)')
axes[1].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
axes[1].set_title('–®–∏—Ä–∏–Ω–∞ –ª–µ–ø–µ—Å—Ç–∫–∞')
axes[1].legend()
axes[1].grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è
print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤:")
print("- –ù–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ –¥–ª–∏–Ω—ã –ª–µ–ø–µ—Å—Ç–∫–∞ –≤–∏–¥–Ω–æ, —á—Ç–æ setosa –ø–æ–ª–Ω–æ—Å—Ç—å—é –æ—Ç–¥–µ–ª–µ–Ω–∞, –Ω–æ versicolor –∏ virginica –ø–µ—Ä–µ–∫—Ä—ã–≤–∞—é—Ç—Å—è.")
print("- –ù–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ —à–∏—Ä–∏–Ω—ã –ª–µ–ø–µ—Å—Ç–∫–∞ —Å–∏—Ç—É–∞—Ü–∏—è –∞–Ω–∞–ª–æ–≥–∏—á–Ω–∞—è.")
print("- –ù–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ –æ—Ç–Ω–æ—à–µ–Ω–∏—è –¥–ª–∏–Ω–∞/—à–∏—Ä–∏–Ω–∞ –≤—Å–µ —Ç—Ä–∏ –≤–∏–¥–∞ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã –ª—É—á—à–µ: setosa –∏–º–µ–µ—Ç —Å–∞–º—ã–µ –≤—ã—Å–æ–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è (‚àº4‚Äì7), versicolor ‚Äì —Å—Ä–µ–¥–Ω–∏–µ (‚àº2.5‚Äì4), virginica ‚Äì –Ω–∏–∑–∫–∏–µ (‚àº1.5‚Äì2.5), –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ.")
print("–í—ã–≤–æ–¥: –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ —É–ª—É—á—à–∞–µ—Ç —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏.")
````

````python
# ========================================
# –®–ê–ì 1: –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏
# ========================================
import numpy as np
np.random.seed(42)  # –§–∏–∫—Å–∏—Ä—É–µ–º —Å–ª—É—á–∞–π–Ω–æ—Å—Ç—å –î–û –≤—Å–µ—Ö –æ–ø–µ—Ä–∞—Ü–∏–π

# ========================================
# –®–ê–ì 2: –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ
# ========================================
from sklearn.datasets import load_wine
import pandas as pd

wine = load_wine()
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df['target'] = wine.target
df['region'] = pd.Categorical.from_codes(wine.target, [f'Region_{i+1}' for i in range(3)])

print("‚úì –î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã. –§–æ—Ä–º–∞:", df.shape)
print("\n–ü–µ—Ä–≤—ã–µ 3 —Å—Ç—Ä–æ–∫–∏:")
display(df.head(3))

# ========================================
# –®–ê–ì 3: –ê–Ω–∞–ª–∏–∑ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
# ========================================
print("\n=== –¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö –∏ –ø—Ä–æ–ø—É—Å–∫–∏ ===")
print(df.info())

print("\n=== –û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ (–ø–µ—Ä–≤—ã–µ 5 –ø—Ä–∏–∑–Ω–∞–∫–æ–≤) ===")
display(df.iloc[:, :5].describe().round(2))

# ========================================
# –®–ê–ì 4: –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π
# ========================================
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white", palette="husl")

# –¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π (—Ç–æ–ª—å–∫–æ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤)
plt.figure(figsize=(12, 10))
corr_matrix = df.iloc[:, :-2].corr()  # –ò—Å–∫–ª—é—á–∞–µ–º target –∏ region
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # –°–∫—Ä—ã–≤–∞–µ–º –≤–µ—Ä—Ö–Ω–∏–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f", cmap='RdBu_r',
            vmin=-1, vmax=1, center=0, square=True, linewidths=.5)
plt.title('–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–∞—Ç–∞—Å–µ—Ç–∞ Wine', fontsize=14, pad=20)
plt.tight_layout()
plt.show()

# ========================================
# –®–ê–ì 5: –ê–Ω–∞–ª–∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –∫–ª—é—á–µ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
# ========================================
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# –ê–ª–∫–æ–≥–æ–ª—å
sns.histplot(data=df, x='alcohol', hue='region', kde=True, ax=axes[0, 0], bins=15)
axes[0, 0].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∞–ª–∫–æ–≥–æ–ª—è –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º')

# –§–ª–∞–≤–æ–Ω–æ–∏–¥—ã
sns.boxplot(data=df, x='region', y='flavanoids', ax=axes[0, 1])
axes[0, 1].set_title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ñ–ª–∞–≤–æ–Ω–æ–∏–¥–æ–≤ (—è—â–∏–∫–∏ —Å —É—Å–∞–º–∏)')

# –ò–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å —Ü–≤–µ—Ç–∞
sns.violinplot(data=df, x='region', y='color_intensity', ax=axes[1, 0])
axes[1, 0].set_title('–ò–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å —Ü–≤–µ—Ç–∞ (–≤–∏–æ–ª–æ–Ω—á–µ–ª—å–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏)')

# –Ø–±–ª–æ—á–Ω–∞—è –∫–∏—Å–ª–æ—Ç–∞
sns.stripplot(data=df, x='region', y='malic_acid', ax=axes[1, 1], jitter=True, size=4)
axes[1, 1].set_title('–Ø–±–ª–æ—á–Ω–∞—è –∫–∏—Å–ª–æ—Ç–∞ (—Ç–æ—á–µ—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫)')

plt.tight_layout()
plt.show()

# ========================================
# –®–ê–ì 6: –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é
# ========================================
X = df.drop(['target', 'region'], axis=1).values
y = df['target'].values

print(f"\n‚úì –ú–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ X: {X.shape}")
print(f"‚úì –í–µ–∫—Ç–æ—Ä –º–µ—Ç–æ–∫ y: {y.shape}")
print(f"‚úì –ë–∞–ª–∞–Ω—Å –∫–ª–∞—Å—Å–æ–≤: {np.bincount(y)}")  # –°–∫–æ–ª—å–∫–æ –æ–±—Ä–∞–∑—Ü–æ–≤ –≤ –∫–∞–∂–¥–æ–º —Ä–µ–≥–∏–æ–Ω–µ
````

### üí° –°–æ–≤–µ—Ç –¥–ª—è –ø—Ä–æ–¥–≤–∏–Ω—É—Ç—ã—Ö —Å—Ç—É–¥–µ–Ω—Ç–æ–≤
–ü–æ–ø—Ä–æ–±—É–π—Ç–µ –ø—Ä–∏–º–µ–Ω–∏—Ç—å **Principal Component Analysis (PCA)** –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö –≤ 2D-–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ. –£–¥–∞—Å—Ç—Å—è –ª–∏ –æ—Ç–¥–µ–ª–∏—Ç—å –≤—Å–µ —Ç—Ä–∏ —Ä–µ–≥–∏–æ–Ω–∞ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç? –≠—Ç–æ –∑–∞–¥–∞–Ω–∏–µ –ø–æ–¥–≥–æ—Ç–æ–≤–∏—Ç –≤–∞—Å –∫ –ø—Ä–∞–∫—Ç–∏–∫–µ 2.1 –ø–æ –ø–æ–Ω–∏–∂–µ–Ω–∏—é —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏.

–£–¥–∞—á–∏ –≤ —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ–π —Ä–∞–±–æ—Ç–µ! üç∑üìä

–ù–∞–π–¥–∏—Ç–µ –≤—Å–µ –ø–∞—Ä—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å –∞–±—Å–æ–ª—é—Ç–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–µ–π |r| > 0.7. –°–æ–∑–¥–∞–π—Ç–µ —Ç–∞–±–ª–∏—Ü—É –∏–∑ 3 —Å—Ç–æ–ª–±—Ü–æ–≤: –ü—Ä–∏–∑–Ω–∞–∫_1, –ü—Ä–∏–∑–Ω–∞–∫_2, –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è. –°–∫–æ–ª—å–∫–æ —Ç–∞–∫–∏—Ö –ø–∞—Ä —Å—É—â–µ—Å—Ç–≤—É–µ—Ç? –ö–∞–∫–æ–π –ø—Ä–∏–∑–Ω–∞–∫ —É—á–∞—Å—Ç–≤—É–µ—Ç –≤ –Ω–∞–∏–±–æ–ª—å—à–µ–º —á–∏—Å–ª–µ —Å–∏–ª—å–Ω—ã—Ö –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

np.random.seed(42)

# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
wine_bunch = load_wine()
wine_df = pd.DataFrame(wine_bunch.data, columns=wine_bunch.feature_names)

# ========== 1. –°—Ç—Ä—É–∫—Ç—É—Ä–Ω—ã–π –∞–Ω–∞–ª–∏–∑ ==========
print("–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö:")
print(wine_df.info())

# ========== 2. –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ ==========
print("\n–û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
print(wine_df.describe().round(3))

# ========== 3. –í–∏–∑—É–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ ‚Äì —Ç–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π ==========
plt.figure(figsize=(11, 9))
sns.heatmap(wine_df.corr(), annot=True, fmt='.3f', cmap='vlag',
            center=0, square=True, linewidths=0.3,
            cbar_kws={'label': '–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏'})
plt.title('–¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤–∏–Ω–∞', fontsize=14, pad=12)
plt.tight_layout()
plt.show()

# ========== –ó–∞–¥–∞–Ω–∏–µ 1: –ü–æ–∏—Å–∫ –ø–∞—Ä —Å |r| > 0.7 ==========
THRESH = 0.7
strong_pairs = []
cols = wine_df.columns
for i in range(len(cols)):
    for j in range(i + 1, len(cols)):
        corr_val = wine_df[cols[i]].corr(wine_df[cols[j]])
        if abs(corr_val) > THRESH:
            strong_pairs.append([cols[i], cols[j], round(corr_val, 4)])

pairs_df = pd.DataFrame(strong_pairs, columns=['–ü—Ä–∏–∑–Ω–∞–∫_1', '–ü—Ä–∏–∑–Ω–∞–∫_2', '–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è'])
pairs_df.sort_values('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è', ascending=False, inplace=True)

print("\n–ü–ê–†–´ –° |r| > 0.7")
print(pairs_df.to_string(index=False))

# ========== –ó–∞–¥–∞–Ω–∏–µ 2: –°–∫–æ–ª—å–∫–æ —Ç–∞–∫–∏—Ö –ø–∞—Ä ==========
num_pairs = len(pairs_df)
print(f"\n–í—Å–µ–≥–æ –ø–∞—Ä: {num_pairs}")

# ========== –ó–∞–¥–∞–Ω–∏–µ 3: –ö–∞–∫–æ–π –ø—Ä–∏–∑–Ω–∞–∫ —É—á–∞—Å—Ç–≤—É–µ—Ç —á–∞—â–µ –≤—Å–µ–≥–æ ==========
all_features = list(pairs_df['–ü—Ä–∏–∑–Ω–∞–∫_1']) + list(pairs_df['–ü—Ä–∏–∑–Ω–∞–∫_2'])
freq = pd.Series(all_features).value_counts()
top_feature = freq.idxmax()
top_count = freq.max()
print(f"\n–ü—Ä–∏–∑–Ω–∞–∫ —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º —Å–≤—è–∑–µ–π: '{top_feature}' ({top_count} –ø–∞—Ä).")

print("\n–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è: –í—ã—Å–æ–∫–∏–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —É–∫–∞–∑—ã–≤–∞—é—Ç –Ω–∞ –∏–∑–±—ã—Ç–æ—á–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —á—Ç–æ –º–æ–∂–µ—Ç –≤–ª–∏—è—Ç—å –Ω–∞ –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∏ PCA. –†–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è —Ä–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å –æ—Ç–±–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–ª–∏ –ø–æ–Ω–∏–∂–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏.")
````

–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∞—Å–∏–º–º–µ—Ç—Ä–∏–∏ (skewness) –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ —Å –ø–æ–º–æ—â—å—é scipy.stats.skew(). –ö–∞–∫–∏–µ 3 –ø—Ä–∏–∑–Ω–∞–∫–∞ –∏–º–µ—é—Ç –Ω–∞–∏–±–æ–ª—å—à—É—é –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—É—é –∞—Å–∏–º–º–µ—Ç—Ä–∏—é (–ø—Ä–∞–≤—ã–π —Ö–≤–æ—Å—Ç)? –ü–æ—á–µ–º—É —ç—Ç–æ –≤–∞–∂–Ω–æ –¥–ª—è –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤, —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã—Ö –∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é (–Ω–∞–ø—Ä–∏–º–µ—Ä, SVM —Å —è–¥—Ä–æ–º RBF)?

````python
# –ò–º–ø–æ—Ä—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
from sklearn.datasets import load_wine

# –§–∏–∫—Å–∏—Ä—É–µ–º –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏
np.random.seed(42)

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –æ –≤–∏–Ω–µ
wine_data = load_wine()
df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# ============================================================
# 1. –°—Ç—Ä—É–∫—Ç—É—Ä–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö
# ============================================================
print("=== –°–¢–†–£–ö–¢–£–†–ù–´–ô –ê–ù–ê–õ–ò–ó (info) ===")
print(df.info())
print("\n")

# ============================================================
# 2. –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑
# ============================================================
print("=== –°–¢–ê–¢–ò–°–¢–ò–ß–ï–°–ö–ò–ô –ê–ù–ê–õ–ò–ó (describe) ===")
print(df.describe().round(3))
print("\n")

# ============================================================
# 3. –í–∏–∑—É–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ ‚Äì —Ç–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π
# ============================================================
plt.figure(figsize=(12, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, square=True, linewidths=0.5,
            cbar_kws={'label': '–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏'})
plt.title('–¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–≤–∏–Ω–æ)', fontsize=14)
plt.tight_layout()
plt.show()

# ============================================================
# 4. –†–∞—Å—á—ë—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∞—Å–∏–º–º–µ—Ç—Ä–∏–∏ (skewness) –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
# ============================================================
skew_values = {}
for col in df.columns:
    skew_values[col] = skew(df[col])

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ Series –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞
skew_series = pd.Series(skew_values).sort_values(ascending=False)
print("\n=== –ö–û–≠–§–§–ò–¶–ò–ï–ù–¢ –ê–°–ò–ú–ú–ï–¢–†–ò–ò –î–õ–Ø –í–°–ï–• –ü–†–ò–ó–ù–ê–ö–û–í ===")
print(skew_series.round(3))

# ============================================================
# 5. –¢—Ä–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞ —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π –∞—Å–∏–º–º–µ—Ç—Ä–∏–µ–π
# ============================================================
positive_skew = skew_series[skew_series > 0]
top3_positive = positive_skew.head(3)
print("\n=== –¢–†–ò –ü–†–ò–ó–ù–ê–ö–ê –° –ù–ê–ò–ë–û–õ–¨–®–ï–ô –ü–û–õ–û–ñ–ò–¢–ï–õ–¨–ù–û–ô –ê–°–ò–ú–ú–ï–¢–†–ò–ï–ô ===")
for feature, val in top3_positive.items():
    print(f"{feature}: {val:.3f}")

# ============================================================
# 6. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π —Ç–æ–ø-3 –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
# ============================================================
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π –∞—Å–∏–º–º–µ—Ç—Ä–∏–µ–π', fontsize=14)

for i, (feature, ax) in enumerate(zip(top3_positive.index, axes)):
    ax.hist(df[feature], bins=20, edgecolor='black', alpha=0.7, color='steelblue')
    ax.axvline(df[feature].mean(), color='red', linestyle='--', label=f'–°—Ä–µ–¥–Ω–µ–µ: {df[feature].mean():.2f}')
    ax.axvline(df[feature].median(), color='green', linestyle=':', label=f'–ú–µ–¥–∏–∞–Ω–∞: {df[feature].median():.2f}')
    ax.set_title(f'{feature}\nskew = {top3_positive[feature]:.3f}')
    ax.set_xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
    ax.set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

–°–∫–æ–Ω—Å—Ç—Ä—É–∏—Ä—É–π—Ç–µ –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ –∫–∞–∫ –ª–∏–Ω–µ–π–Ω—É—é –∫–æ–º–±–∏–Ω–∞—Ü–∏—é –¥–≤—É—Ö —Å—É—â–µ—Å—Ç–≤—É—é—â–∏—Ö:

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —ç—Ç–æ–≥–æ –∏–Ω–¥–µ–∫—Å–∞ –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º. –£–ª—É—á—à–∞–µ—Ç –ª–∏ –æ–Ω —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏? –ü–æ–¥—Ç–≤–µ—Ä–¥–∏—Ç–µ –≤—ã–≤–æ–¥ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ —á–µ—Ä–µ–∑ —Ä–∞—Å—á—ë—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤ (—Ä–∞–∑–Ω–∏—Ü–∞ —Å—Ä–µ–¥–Ω–∏—Ö / —Å—É–º–º–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π).

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

np.random.seed(42)

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target
df['cultivar'] = df['target'].map({0: '—Å–æ—Ä—Ç_1', 1: '—Å–æ—Ä—Ç_2', 2: '—Å–æ—Ä—Ç_3'})

# ============================================================
# 1. –°—Ç—Ä—É–∫—Ç—É—Ä–Ω—ã–π –∞–Ω–∞–ª–∏–∑
# ============================================================
print("–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö:")
print(df.info())
print("\n")

# ============================================================
# 2. –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑
# ============================================================
print("–û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
print(df[wine.feature_names].describe().round(3))
print("\n")

# ============================================================
# 3. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è ‚Äì —Ç–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π
# ============================================================
plt.figure(figsize=(12, 10))
corr = df[wine.feature_names].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=0.5, cbar_kws={'label': '–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è'})
plt.title('–¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤–∏–Ω–∞')
plt.tight_layout()
plt.show()

# ============================================================
# –ó–ê–î–ê–ù–ò–ï: –ö–æ–Ω—Å—Ç—Ä—É–∏—Ä–æ–≤–∞–Ω–∏–µ –Ω–æ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –∏ –æ—Ü–µ–Ω–∫–∞ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è
# ============================================================
print("="*60)
print("–ö–û–ù–°–¢–†–£–ò–†–û–í–ê–ù–ò–ï –ù–û–í–û–ì–û –ü–†–ò–ó–ù–ê–ö–ê")
print("="*60)

# –°–æ–∑–¥–∞–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞
df['wine_quality_index'] = 0.6 * df['flavanoids'] + 0.4 * df['color_intensity']
print("–°–æ–∑–¥–∞–Ω –ø—Ä–∏–∑–Ω–∞–∫ 'wine_quality_index' = 0.6*flavanoids + 0.4*color_intensity")

# –ì—Ä–∞—Ñ–∏–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ —Å–æ—Ä—Ç–∞–º
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞
for cult in df['cultivar'].unique():
    subset = df[df['cultivar'] == cult]
    axes[0].hist(subset['wine_quality_index'], bins=12, alpha=0.6, label=cult, edgecolor='black')
axes[0].set_xlabel('wine_quality_index')
axes[0].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
axes[0].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞ –ø–æ —Å–æ—Ä—Ç–∞–º')
axes[0].legend()
axes[0].grid(alpha=0.3)

# –Ø—â–∏–∫ —Å —É—Å–∞–º–∏
df.boxplot(column='wine_quality_index', by='cultivar', ax=axes[1])
axes[1].set_title('–Ø—â–∏–∫ —Å —É—Å–∞–º–∏ –¥–ª—è –∏–Ω–¥–µ–∫—Å–∞')
axes[1].set_xlabel('–°–æ—Ä—Ç')
axes[1].set_ylabel('wine_quality_index')
axes[1].grid(alpha=0.3)
plt.suptitle('')
plt.tight_layout()
plt.show()

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á—ë—Ç–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è
def sep_coef(g1, g2):
    return abs(g1.mean() - g2.mean()) / (g1.std() + g2.std() + 1e-10)

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å –∏—Å—Ö–æ–¥–Ω—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏
features = ['flavanoids', 'color_intensity', 'wine_quality_index']
cultivars = sorted(df['cultivar'].unique())
results = []

print("\n–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –º–µ–∂–¥—É –ø–∞—Ä–∞–º–∏ —Å–æ—Ä—Ç–æ–≤:")
for feat in features:
    print(f"\n{feat}:")
    for i in range(len(cultivars)):
        for j in range(i+1, len(cultivars)):
            c1, c2 = cultivars[i], cultivars[j]
            d1 = df[df['cultivar'] == c1][feat]
            d2 = df[df['cultivar'] == c2][feat]
            val = sep_coef(d1, d2)
            results.append((feat, f'{c1}-{c2}', round(val, 3)))
            print(f"  {c1} vs {c2}: {val:.3f}")

# –°—Ä–µ–¥–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã
print("\n–°—Ä–µ–¥–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è:")
avg_dict = {}
for feat in features:
    vals = [r[2] for r in results if r[0] == feat]
    avg_dict[feat] = sum(vals) / len(vals) if vals else 0
    print(f"  {feat}: {avg_dict[feat]:.3f}")

best = max(avg_dict, key=avg_dict.get)
print(f"\n–õ—É—á—à–∏–π –ø—Ä–∏–∑–Ω–∞–∫ –ø–æ —Ä–∞–∑–¥–µ–ª—è—é—â–µ–π —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç–∏: {best} ({avg_dict[best]:.3f})")
if best == 'wine_quality_index':
    print("–í—ã–≤–æ–¥: –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ —É–ª—É—á—à–∞–µ—Ç —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ —Å–æ—Ä—Ç–æ–≤ –≤–∏–Ω–∞.")
else:
    print("–í—ã–≤–æ–¥: –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ –Ω–µ —É–ª—É—á—à–∞–µ—Ç —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ; –ª—É—á—à–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏—Å—Ö–æ–¥–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏.")

print("\n–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è: –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Ö–æ—Ä–æ—à–æ –ø—Ä–∏–∑–Ω–∞–∫ —Ä–∞–∑–ª–∏—á–∞–µ—Ç –∫–ª–∞—Å—Å—ã. –ß–µ–º –≤—ã—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ, —Ç–µ–º –ª—É—á—à–µ. –ê–Ω–∞–ª–∏–∑ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç (–∏–ª–∏ –æ–ø—Ä–æ–≤–µ—Ä–≥–∞–µ—Ç) –ø–æ–ª–µ–∑–Ω–æ—Å—Ç—å —Å–∫–æ–Ω—Å—Ç—Ä—É–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –∏–Ω–¥–µ–∫—Å–∞.")
````
